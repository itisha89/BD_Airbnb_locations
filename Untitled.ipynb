This program will print the DataFrame df with a new column named 'dates_2022', containing dates from both columns ('first_review' and 'last_review') that occurred in the year 2022. If the year is 2022, it concatenates both dates into a single string in the format "DD-MM-YYYY, DD-MM-YYYY" and puts it in the new column 'dates_2022'.

```python
# Convert string dates to datetime format
df['first_review'] = pd.to_datetime(df['first_review'], format='%Y-%m-%d')
df['last_review'] = pd.to_datetime(df['last_review'], format='%Y-%m-%d')
```

```python
# Convert string dates to datetime format


# Create a new column 'dates_2022' containing dates from 'first_review' and 'last_review' that are in the year 2022
df['dates_2022'] = df.apply(lambda row: row['first_review'].strftime('%d-%m-%Y') + ', ' + row['last_review'].strftime('%d-%m-%Y') if row['first_review'].year == 2022 or row['last_review'].year == 2022 else 'Null', axis=1)

# Print the DataFrame with the new column
print("DataFrame with 'dates_2022' column:")
df.head()
```
